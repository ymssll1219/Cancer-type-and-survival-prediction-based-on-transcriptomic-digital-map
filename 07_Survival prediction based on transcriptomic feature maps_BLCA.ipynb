```python
from keras import models
from keras import layers
from keras.models import Sequential,Input
from keras import optimizers
import keras
from keras.layers import Conv2D,MaxPooling2D,Dropout,BatchNormalization,Flatten,Dense,Lambda,Multiply,AveragePooling2D,Activation,Add,LeakyReLU
from keras.models import Model
```

```python
def Conv2d_BN(x, nb_filter, kernel_size, strides=(1,1), padding='same'):
    x = Conv2D(nb_filter, kernel_size, strides=strides, padding=padding)(x)
    x = BatchNormalization(axis=3)(x)
    x = LeakyReLU(alpha=0.1)(x)
    return x
```

```python
def TestNet(input_shape=None):
    inpt = Input(shape=input_shape)
    #padding = 'same'，填充为(步长-1）/2,还可以用ZeroPadding3D((3,3,3))
    x = Conv2d_BN(inpt, 64, (5, 5))
    x = Conv2d_BN(x, 64, (5, 5))
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2),padding='same')(x)
    x = Conv2d_BN(x, 128, (3, 3))
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2),padding='same')(x)
    x = Conv2d_BN(x, 128, (3, 3))
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2),padding='same')(x)
    x = Conv2d_BN(x, 128, (3, 3))
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2),padding='same')(x)
    #x = Dropout(0.3)(x)
    x = Flatten()(x)
    x = Dense(128,activation='relu')(x)
    x = Dropout(0.1)(x)
    x = Dense(1,activation='sigmoid')(x)
    model = Model(inpt,x,name='inception')
    
 
    return model
```

```python
model = TestNet(input_shape=(33,33,1))
model.summary()
```

```python
from tensorflow.keras.utils import plot_model     ##模型结构图
plot_model(model, 'd:\\h5\\model_3d.png')
```

```python
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=2)
cvscores = []
for train, test in kfold.split(imgm, lab):
    model = TestNet(input_shape=(33,33,1))
    model.compile(loss='binary_crossentropy',
              optimizer=optimizers.Adam(lr=1e-4),#
              metrics=['acc'])
    model.fit(imgm[train], lab[train], epochs=100, batch_size=64, verbose=0)
    scores = model.evaluate(imgm[test], lab[test], verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)
```

```python
Data_X_min, Data_X_max = np.min(img), np.max(img)
Data_X_mean = np.mean(img)  
```

```python
def normalize(array):
    #array = (array - Data_X_min) / (Data_X_max - Data_X_min)
    array = (array - Data_X_mean) / (Data_X_max - Data_X_min)
    return array
```

```python
img_n = normalize(img)
```

```python
img_n = img_n.reshape(len(img), 33,33,1)
```

```python
img_n.min()
```

```python
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)
cvscores = []
for train, test in kfold.split(img_n, lab):
    model1 = TestNet(input_shape=(33,33,1))
    model1.compile(loss='binary_crossentropy',
              optimizer=optimizers.Adam(lr=1e-4),#
              metrics=['acc'])
    X_traingan = np.vstack((img_n[train],img0,img1))
    y_traingan = np.concatenate((lab[train],lab0,lab1))
    model1.fit(X_traingan, y_traingan, epochs=100, batch_size=64, verbose=0)
    scores = model1.evaluate(img_n[test], lab[test], verbose=0)
    print("%s: %.2f%%" % (model1.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)
```
