```python
def tran(N,data,i):
    myarray = np.zeros((N, N))
    df = np.zeros((N, N))
    x, y = 0, N - 1
    res = myarray[x][y] = 1

    while res < N * N:
        # 向下移动
        while x + 1 < N and not myarray[x + 1][y]:
            res += 1
            x += 1
            myarray[x][y] = res
            df[x][y] = data[i][min(res - 1, 1088)]

        # 向左移动
        while y - 1 >= 0 and not myarray[x][y - 1]:
            res += 1
            y -= 1
            myarray[x][y] = res
            df[x][y] = data[i][min(res - 1, 1088)]

        # 向上移动
        while x - 1 >= 0 and not myarray[x - 1][y]:
            res += 1
            x -= 1
            myarray[x][y] = res
            df[x][y] = data[i][min(res - 1, 1088)]

        # 向右移动
        while y + 1 < N and not myarray[x][y + 1]:
            res += 1
            y += 1
            myarray[x][y] = res
            df[x][y] = data[i][min(res - 1, 1088)]
    return df
```

```python
df = tran(33,data,len(feat_x))
df.shape
```
