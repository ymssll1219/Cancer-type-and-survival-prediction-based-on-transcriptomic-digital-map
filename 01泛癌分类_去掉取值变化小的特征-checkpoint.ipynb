```python
import pandas as pd
patho = r'01_genenametq.tsv'
data0 = pd.read_table(patho,header=None,delim_whitespace=True)#encoding='gb2312',index_col=0
print(data0.shape)
names1 = data0[0]
names2 = data0[1]
names1 = names1.values.tolist()
names2 = names2.values.tolist()
print(len(names1))
```

```python
import time
from tqdm import tqdm
from tqdm._tqdm import trange
```

```python
blist1 = []
qlist1 = []
blist2 = []
qlist2 = []
for i in tqdm(range(len(names1))):
    time.sleep(0.01)
    a = std[i]
    if a <= 1:
        qlist1.append(i)
        qlist2.append(i)
    else:
        blist1.append(i)
        blist2.append(i)
        #feature = np.delete(feature,i,1)
        #del names[i]
    #print(names[i])
```

```python
blist1 = np.array(blist1)
qlist1 = np.array(qlist1)
blist2 = np.array(blist2)
qlist2 = np.array(qlist2)
```
